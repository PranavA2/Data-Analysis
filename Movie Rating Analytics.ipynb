```python
list1 = []
list2 = []
for gen in movies.Genre.cat.categories:
    list1.append(movies[movies.Genre == gen].BudgetMillion)
    list2.append(gen)

sns.set_style("whitegrid")
fig, ax = plt.subplots()
fig.set_size_inches(11.7, 8.27)
h = plt.hist(list1,bins=30,stacked=True,rwidth=1, label=list2)
plt.title("Movie Budget Distribution", fontsize=20, color="DarkBlue", fontname="Arial")
plt.ylabel("Number of Movies", fontsize=20, color="Red",fontname="Arial")
plt.xlabel("Budget",fontsize=20,color="Green",fontname="Arial")
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.legend(prop={'size':20},fancybox=True,shadow=True,framealpha=1)
plt.show()
```
